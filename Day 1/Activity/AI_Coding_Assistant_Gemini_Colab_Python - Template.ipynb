# **AI Coding Assistant Activity**

## Objective

Learn how to use **Google Gemini** as an AI coding assistant to accelerate basic Python development while understanding the generated code.

**Duration:** 5-10 minutes

**Platform:** Google Colab

---

## **Part 1. Variables and Basic Calculations**

**Task**  

Use Gemini to generate Python code that:  
- Creates variables for your name, age, and favorite programming language.
- Prints a short self-introduction.
- Calculates your age after 10 years.

**Example Prompt**  

```md
Write Python code that:
- Creates variables for my name, age, and favorite programming language.
- Prints a friendly self-introduction.
- Calculates and prints my age after 10 years.
Use beginner-friendly Python with comments.
```

**Challenge**

Modify the generated code so that it also prints:
- Birth year (assume the current year is 2026)
- Number of years until age 100

## **Part 2. Lists and Loops**

**Task**  

Use Gemini to generate code that works with a list of exam scores.

**Example scores:**

```python
scores = [85, 72, 91, 68, 95]
```

The program should:
- Print every score
- Calculate the average score
- Display the highest and lowest scores

**Example Prompt**

```md
Write Python code that uses a list of exam scores.

The program should:
- Print every score using a loop.
- Calculate the average.
- Print the maximum and minimum scores.

Explain each step with comments.
```

**Challenge**

Ask Gemini to modify the program so it also prints:
- Number of students
- Number of scores above 80

## **Part 3. Functions**

**Task**  

Create a reusable function using Gemini.

The function should:
- Accept a student's score
- Return the corresponding grade

**Example grading scheme:**

| Score | Grade |
| ----- | ----- |
| ≥ 80  | A     |
| ≥ 70  | B     |
| ≥ 60  | C     |
| ≥ 50  | D     |
| < 50  | F     |

**Example Prompt**

```md
Write a Python function called get_grade(score).

The function should:
- Return A, B, C, D, or F based on the score.
- Include comments.
- Show example function calls.
```

**Challenge**

Ask Gemini to improve the program by:
- Testing multiple scores stored in a list
- Printing a formatted report showing both the score and grade

## Reflection

Discuss with your partner:
- Which parts of the code did Gemini generate correctly?
- Did you understand every line of the generated code?
- What changes did you make after receiving Gemini's answer?
- When is it useful to rely on AI coding assistance, and when should you verify the output yourself?
