## **Part 2. Lists and Loops**

**Task**  

Use Gemini to generate code that works with a list of exam scores.

**Example scores:**

```python
scores = [85, 72, 91, 68, 95]
```

The program should:
- Print every score
- Calculate the average score
- Display the highest and lowest scores

**Example Prompt**

```md
Write Python code that uses a list of exam scores.

The program should:
- Print every score using a loop.
- Calculate the average.
- Print the maximum and minimum scores.

Explain each step with comments.
```

**Challenge**

Ask Gemini to modify the program so it also prints:
- Number of students
- Number of scores above 80

````python
scores = [85, 72, 91, 68, 95]

# Print every score using a loop
print("Exam Scores:")
for score in scores:
    print(score)

# Calculate the average score
total_score = sum(scores)
number_of_scores = len(scores)
if number_of_scores > 0:
    average_score = total_score / number_of_scores
    print(f"\nAverage Score: {average_score:.2f}")
else:
    print("\nNo scores to calculate average.")

# Display the highest and lowest scores
if number_of_scores > 0:
    highest_score = max(scores)
    lowest_score = min(scores)
    print(f"Highest Score: {highest_score}")
    print(f"Lowest Score: {lowest_score}")
else:
    print("Cannot determine highest and lowest scores from an empty list.")
````

## **Part 3. Functions**

**Task**  

Create a reusable function using Gemini.

The function should:
- Accept a student's score
- Return the corresponding grade

**Example grading scheme:**

| Score | Grade |
| ----- | ----- |
| ≥ 80  | A     |
| ≥ 70  | B     |
| ≥ 60  | C     |
| ≥ 50  | D     |
| < 50  | F     |

**Example Prompt**

```md
Write a Python function called get_grade(score).

The function should:
- Return A, B, C, D, or F based on the score.
- Include comments.
- Show example function calls.
```

**Challenge**

Ask Gemini to improve the program by:
- Testing multiple scores stored in a list
- Printing a formatted report showing both the score and grade

````python
# Define a function called get_grade that takes a 'score' as input
def get_grade(score):
    # Check the score against the grading scheme
    if score >= 80:
        return "A"
    elif score >= 70:
        return "B"
    elif score >= 60:
        return "C"
    elif score >= 50:
        return "D"
    else:
        return "F"

# --- Example Function Calls ---
print("--- Example Grades ---")

# Example 1: Score that should return 'A'
score1 = 85
grade1 = get_grade(score1)
print(f"A score of {score1} gets a grade: {grade1}")

# Example 2: Score that should return 'B'
score2 = 72
grade2 = get_grade(score2)
print(f"A score of {score2} gets a grade: {grade2}")

# Example 3: Score that should return 'F'
score3 = 45
grade3 = get_grade(score3)
print(f"A score of {score3} gets a grade: {grade3}")

# Example 4: Edge case score for 'A'
score4 = 80
grade4 = get_grade(score4)
print(f"A score of {score4} gets a grade: {grade4}")

# Example 5: Edge case score for 'D'
score5 = 50
grade5 = get_grade(score5)
print(f"A score of {score5} gets a grade: {grade5}")
````
