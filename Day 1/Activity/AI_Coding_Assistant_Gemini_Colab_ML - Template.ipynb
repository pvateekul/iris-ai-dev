## Part 1 - Python & Pandas

### Objective

Learn how to use Gemini as an AI coding assistant to perform exploratory data analysis (EDA) and basic data preprocessing using Pandas.

### Step 1: Explore the Dataset

You already have a pandas DataFrame named `df`.

**Prompt Gemini:**

> I have a pandas DataFrame called `df`.
>
> Help me write Python code to:
>
> - "Display the first five rows."
> - "Show the dataset shape."
> - "List all column names."
> - "Display the data type of each column."
>
> Briefly explain what each command does.

**After running the generated code:**

- Are the columns what you expected?
- How many rows and columns are in the dataset?

### Step 2: Assess Data Quality

**Prompt Gemini:**

> Using the DataFrame `df`, help me inspect the dataset quality.
>
> Generate Python code to:
>
> * Check for missing values.
> * Count duplicate rows.
> * Display summary statistics.
> * Show the number of unique values in each categorical column.
>
> Explain what each result tells us.

**After running the generated code:**

* Does the dataset contain missing values?
* Are there duplicate records?
* Which columns are categorical?

### Step 3: Clean the Dataset

**Prompt Gemini:**

> Help me clean the dataset.
>
> Generate Python code to:
>
> * Remove duplicate rows.
> * Handle missing values using an appropriate strategy.
> * Verify that the cleaned dataset contains no duplicate rows.
>
> Briefly explain why each preprocessing step is important.

**After running the generated code:**

* Did the number of rows change?
* Were all missing values handled correctly?

### Step 4: Feature Engineering

**Prompt Gemini:**

> Suggest one useful feature that can be created from this dataset.
>
> Generate the Python code to create the new feature and explain why it could be useful for machine learning.

If Gemini's suggestion is not appropriate for the dataset, ask it to propose an alternative feature.

## Part 2 - Traditional Machine Learning

### Objective

Learn how to use Gemini as an AI coding assistant to build, evaluate, and compare machine learning classification models using Scikit-learn.

### Step 1: Prepare the Dataset

You have a cleaned pandas DataFrame named `df`.

**Prompt Gemini:**

> I have a pandas DataFrame called `df` for a binary classification problem.
>
> Help me prepare the dataset for Scikit-learn by:
>
> * Separating the features (`X`) and target (`y`).
> * Encoding categorical variables if necessary.
> * Splitting the dataset into training and testing sets using an 80/20 split.
>
> Briefly explain why each step is required.

**After running the generated code:**

* How many training and testing samples were created?
* Were any categorical features encoded?

### Step 2: Train a Random Forest Classifier

**Prompt Gemini:**

> Help me build a Scikit-learn classification pipeline using:
>
> * StandardScaler (only if appropriate)
> * RandomForestClassifier
>
> Train the model using the training data and generate predictions on the test set.
>
> Briefly explain each step.

**After running the generated code:**

* Did the model train successfully?
* Which hyperparameters were used?

### Step 3: Evaluate the Model

**Prompt Gemini:**

> Generate Python code to evaluate my classification model using:
>
> * Accuracy
> * Precision
> * Recall
> * F1-score
> * Confusion Matrix
> * Classification Report
>
> Briefly explain what each evaluation metric measures.

**After running the generated code:**

* Which metric is the most informative for this dataset?
* Is the model performing well?

### Step 4: Improve the Model

**Prompt Gemini:**

> Suggest three hyperparameters of RandomForestClassifier that could improve model performance.
>
> Generate Python code to train the model using your suggested values.

Run the updated model and compare the evaluation metrics with the previous version.

### Challenge: Compare Multiple Models

Ask Gemini to replace the Random Forest classifier with each of the following algorithms:

* Logistic Regression
* Support Vector Machine (SVM)

Evaluate each model using the same train/test split.

Create a comparison table containing:

Finally, ask Gemini:

> Based on these evaluation metrics, which model would you recommend for this dataset? Explain your reasoning.

## Best Practices

* Treat AI as a coding assistant, not a replacement for understanding.
* Read and verify all generated code before running it.
* Test outputs using small examples.
* Ask follow-up questions when code is unclear.
* Compare AI-generated solutions with your own implementation.
