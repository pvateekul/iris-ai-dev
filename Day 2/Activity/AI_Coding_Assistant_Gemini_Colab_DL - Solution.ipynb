```python
from IPython.display import display, Markdown

# Define
markdown_text = """
### What CIFAR-10 is

The **CIFAR-10 dataset** is a very popular dataset in computer vision for training machine learning models, especially for image classification. It consists of **60,000 tiny 32x32 pixel color images**, divided into 10 different classes (e.g., airplane, automobile, bird, cat, deer, dog, frog, horse, ship, truck). There are 6,000 images per class. 50,000 of these images are used for training your model, and the remaining 10,000 are used for testing.

It's a great dataset for beginners because the images are small and the number of classes is manageable, allowing for faster experimentation with different model architectures.

### Why CNNs are suitable for this task

**Convolutional Neural Networks (CNNs)** are particularly well-suited for image classification tasks like CIFAR-10 for a few key reasons:

1.  **Feature Learning**: CNNs can automatically learn important features directly from raw image pixels, rather than needing humans to hand-design features. For example, they can learn to recognize edges, textures, and ultimately more complex patterns like eyes or wheels.
2.  **Spatial Hierarchies**: Images have a spatial structure (pixels are arranged in a grid). CNNs are designed to respect this structure. They use "convolutional layers" that look at small regions of an image and slide across it, detecting patterns regardless of where they appear in the image.
3.  **Parameter Sharing**: Unlike traditional neural networks where every neuron is connected to every input, CNNs use parameter sharing. This means the same set of weights (filters) is applied across different parts of the image, significantly reducing the number of parameters the model needs to learn. This makes them more efficient and helps them generalize better.
4.  **Translation Invariance**: Because CNNs use filters that slide across the image, they become somewhat invariant to the exact position of an object. If a cat is slightly to the left or right, a CNN can still recognize it.

### What a basic CNN architecture looks like

A basic CNN architecture typically involves a sequence of layers designed to process images. Here's a common, beginner-friendly structure:

1.  **Input Layer**: This is where your image data (e.g., a 32x32x3 CIFAR-10 image, where 3 is for RGB color channels) enters the network.

2.  **Convolutional Layer(s) (`Conv2D`)**: These are the core building blocks. Each convolutional layer applies a set of learnable filters (small matrices) to the input image. These filters slide over the image, performing a dot product, and creating a "feature map" that highlights certain patterns (like edges or corners) in the image. You'll often see multiple convolutional layers stacked.

3.  **Activation Function (e.g., `ReLU`)**: After a convolutional operation, an activation function like Rectified Linear Unit (ReLU) is usually applied. ReLU introduces non-linearity into the model, allowing it to learn more complex patterns. It essentially replaces negative values in the feature map with zero.

4.  **Pooling Layer(s) (e.g., `MaxPooling`)**: Pooling layers reduce the spatial dimensions (width and height) of the feature maps, which helps to reduce computational cost and control overfitting. `MaxPooling` takes the maximum value from a small window of the feature map.

5.  **Flatten Layer**: After several convolutional and pooling layers, the 2D feature maps are flattened into a single, long 1D vector. This is necessary to feed the data into the fully connected layers.

6.  **Fully Connected Layer(s) (`Dense`)**: These are standard neural network layers where every neuron is connected to every neuron in the previous layer. They take the high-level features extracted by the convolutional layers and use them to make final predictions.

7.  **Output Layer**: The final fully connected layer typically has a number of neurons equal to the number of classes (10 for CIFAR-10). An activation function like `softmax` is often used here to output probabilities for each class, indicating how likely the image belongs to each class.
"""

# Render and display the markdown
display(Markdown(markdown_text))
```

```python
import torch
import torch.nn as nn
import pytorch_lightning as pl
from torch.utils.data import DataLoader, Subset
from torchvision import datasets, transforms

# -----------------------------
# 1. CNN Model
# -----------------------------
class SimpleCNN(pl.LightningModule):
    def __init__(self):
        super().__init__()

        self.features = nn.Sequential(
            nn.Conv2d(3, 16, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )

        self.classifier = nn.Linear(32 * 8 * 8, 10)

        self.loss_fn = nn.CrossEntropyLoss()

    def forward(self, x):
        x = self.features(x)
        x = torch.flatten(x, 1)
        return self.classifier(x)

    def training_step(self, batch, batch_idx):
        x = batch["pixel_values"]
        y = batch["label"]

        logits = self(x)
        loss = self.loss_fn(logits, y)
        return loss

    def test_step(self, batch, batch_idx):
        x = batch["pixel_values"]
        y = batch["label"]

        logits = self(x)
        loss = self.loss_fn(logits, y)

        preds = logits.argmax(dim=1)
        acc = (preds == y).float().mean()

        self.log("test_loss", loss)
        self.log("test_acc", acc)

    def validation_step(self, batch, batch_idx):
        x, y = batch
        logits = self(x)
        loss = self.loss_fn(logits, y)

        acc = (logits.argmax(1) == y).float().mean()

        self.log("val_loss", loss)
        self.log("val_acc", acc)

    def configure_optimizers(self):
        return torch.optim.Adam(self.parameters(), lr=1e-3)
```

```python
from datasets import load_dataset
from torchvision import transforms
from torch.utils.data import DataLoader

# -----------------------------
# 2. Dataset
# -----------------------------
transform = transforms.ToTensor()

# Load dataset
dataset = load_dataset("uoft-cs/cifar10")

# Small subsets
train_dataset = dataset["train"].select(range(200))
test_dataset = dataset["test"].select(range(100))

# Apply transforms
def preprocess(examples):
    return {
        "pixel_values": [transform(img) for img in examples["img"]],
        "label": examples["label"],
    }

train_dataset = train_dataset.with_transform(preprocess)
test_dataset = test_dataset.with_transform(preprocess)

train_loader = DataLoader(
    train_dataset,
    batch_size=128,
    shuffle=True,
)

test_loader = DataLoader(
    test_dataset,
    batch_size=128,
    shuffle=False,
)
```

```python
# -----------------------------
# 3. Train
# -----------------------------
model = SimpleCNN()

trainer = pl.Trainer(
    max_epochs=1,
    logger=False,
    enable_checkpointing=False,
    enable_progress_bar=False,
)

trainer.fit(model, train_loader)

trainer.test(model, test_loader)
```

```python
markdown_text = """### Explanation of Enhancements and their Benefits

#### 1. Data Augmentation (Random Flips, Crops, Normalization)
*   **How it helps**: Data augmentation artificially increases the size and diversity of your training dataset by applying random transformations (like horizontal flips and small crops) to the input images. Normalization scales pixel values to a standard range (e.g., 0-1 or mean 0, std 1).
*   **Benefit**: By showing the model slightly varied versions of the same image, it learns to be more robust to minor changes in object position, orientation, and lighting. This reduces overfitting, as the model doesn't just memorize the training examples but learns more generalizable features. Normalization helps stabilize and speed up the training process by ensuring consistent input scales across features.

#### 2. Dropout Layers
*   **How it helps**: Dropout is a regularization technique where, during training, a random selection of neurons (and their connections) are temporarily ignored (set to zero) with a given probability. This forces the network to learn more robust features by preventing neurons from co-adapting too much.
*   **Benefit**: Dropout significantly reduces overfitting. Each training step, the model trains a slightly different network, preventing complex co-dependencies on specific neurons. It acts like training an ensemble of many different networks, leading to a more generalized and less sensitive model at inference time.

#### 3. Batch Normalization
*   **How it helps**: Batch Normalization layers are inserted in the network (typically after convolutional layers and before activation functions). They normalize the output of a preceding layer by re-centering and re-scaling the activations in each mini-batch during training. This addresses the problem of "internal covariate shift" where the distribution of layer inputs changes during training.
*   **Benefit**: Batch Normalization stabilizes and accelerates the training process. It allows for higher learning rates, making the model converge faster. It also acts as a form of regularization, reducing the need for strong dropout and further combating overfitting by making the network less sensitive to the initial weights.

#### 4. Early Stopping
*   **How it helps**: Early stopping is a callback that monitors a chosen metric (e.g., validation loss) and stops the training process if the metric stops improving for a specified number of epochs (`patience`).
*   **Benefit**: This is a powerful regularization technique that prevents the model from overfitting. Training for too many epochs can lead the model to memorize the training data and perform poorly on unseen data. Early stopping automatically finds the optimal number of training epochs, saving computational resources and leading to a better generalization performance by stopping when the model's performance on unseen data starts to degrade.
"""

# Render and display the markdown
display(Markdown(markdown_text))
```

```python
import torch
import torch.nn as nn
import pytorch_lightning as pl
from torch.utils.data import DataLoader, Subset
from torchvision import datasets, transforms
from pytorch_lightning.callbacks.early_stopping import EarlyStopping
from pytorch_lightning.loggers import TensorBoardLogger
import os
import pandas as pd
import matplotlib.pyplot as plt
import glob

# ----------------------------------------------------
# 1. Data Augmentation and Normalization
# ----------------------------------------------------
# Define transformations for training data (with augmentation)
# Normalization values are specific to CIFAR-10
train_transform_augmented = transforms.Compose([
    transforms.RandomHorizontalFlip(), # Randomly flip the image horizontally
    transforms.RandomCrop(32, padding=4), # Randomly crop with padding
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2471, 0.2435, 0.2616)) # CIFAR-10 mean and std
])

# Define transformations for test/validation data (only normalization)
test_transform_normalized = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2471, 0.2435, 0.2616))
])

# ----------------------------------------------------
# 2. Dataset and DataLoaders (using FakeData for quick demo)
# ----------------------------------------------------
print("--- Setting up DataLoaders with Augmented Data ---")
# For demonstration purposes, we'll continue using FakeData
# with the new transforms. In a real scenario, you'd use CIFAR10.

train_dataset_improved = datasets.FakeData(
    size=200, # Keep small for fast execution
    image_size=(3, 32, 32),
    num_classes=10,
    transform=train_transform_augmented,
)

test_dataset_improved = datasets.FakeData(
    size=50, # Keep small for fast execution
    image_size=(3, 32, 32),
    num_classes=10,
    transform=test_transform_normalized,
)

train_loader_improved = DataLoader(
    train_dataset_improved,
    batch_size=32, # Smaller batch size for potentially better generalization
    shuffle=True,
    num_workers=0
)

test_loader_improved = DataLoader(
    test_dataset_improved,
    batch_size=32,
    shuffle=False,
    num_workers=0
)

# ----------------------------------------------------
# 3. Improved CNN Model with Batch Normalization and Dropout
# ----------------------------------------------------
class ImprovedCNN(pl.LightningModule):
    def __init__(self):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1), # Increased filters
            nn.BatchNorm2d(32), # Batch Normalization
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout(0.25), # Dropout after pooling

            nn.Conv2d(32, 64, kernel_size=3, padding=1), # Increased filters
            nn.BatchNorm2d(64), # Batch Normalization
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout(0.25) # Dropout after pooling
        )
        # Calculate the size of the flattened output after conv/pool layers
        # For 32x32 input, two MaxPool2d(2) layers reduce it to 8x8
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64 * 8 * 8, 512), # Added a hidden layer
            nn.BatchNorm1d(512), # Batch Normalization for FC layer
            nn.ReLU(),
            nn.Dropout(0.5), # Higher dropout for fully connected layer
            nn.Linear(512, 10)
        )
        self.loss_fn = nn.CrossEntropyLoss()

    def forward(self, x):
        x = self.features(x)
        return self.classifier(x)

    def training_step(self, batch, batch_idx):
        x, y = batch
        logits = self(x)
        loss = self.loss_fn(logits, y)
        preds = logits.argmax(dim=1)
        acc = (preds == y).float().mean()
        self.log("train_loss", loss, on_step=False, on_epoch=True)
        self.log("train_acc", acc, on_step=False, on_epoch=True)
        return loss

    def validation_step(self, batch, batch_idx):
        x, y = batch
        logits = self(x)
        loss = self.loss_fn(logits, y)
        preds = logits.argmax(1)
        acc = (preds == y).float().mean()
        self.log("val_loss", loss, on_step=False, on_epoch=True)
        self.log("val_acc", acc, on_step=False, on_epoch=True)
        return loss

    def test_step(self, batch, batch_idx):
        x, y = batch
        logits = self(x)
        loss = self.loss_fn(logits, y)
        preds = logits.argmax(1)
        acc = (preds == y).float().mean()
        self.log("test_loss", loss)
        self.log("test_acc", acc)

    def configure_optimizers(self):
        return torch.optim.Adam(self.parameters(), lr=1e-3)

# ----------------------------------------------------
# 4. Training with Early Stopping
# ----------------------------------------------------
print("\n--- Training Improved Model with Early Stopping ---")
model_improved = ImprovedCNN()

# Setup TensorBoardLogger for the improved model
log_dir_path_improved = os.path.join("tb_logs", "cifar10_cnn_improved_log")
logger_improved = TensorBoardLogger(save_dir=log_dir_path_improved)

# Early Stopping callback
early_stop_callback = EarlyStopping(
    monitor='val_loss', # Monitor validation loss
    min_delta=0.00, # Minimum change to qualify as an improvement
    patience=3, # Number of epochs with no improvement after which training will be stopped
    verbose=True,
    mode='min'
)

trainer_improved = pl.Trainer(
    max_epochs=10, # Allow more epochs, early stopping will manage it
    logger=logger_improved,
    callbacks=[early_stop_callback],
    enable_checkpointing=False,
    enable_progress_bar=True, # Enable progress bar to see early stopping in action
)

# Train the improved model
trainer_improved.fit(model_improved, train_loader_improved, test_loader_improved) # test_loader as val_loader

# Evaluate the improved model on the test set
print("\n--- Evaluating Improved Model Test Accuracy ---")
trainer_improved.test(model_improved, test_loader_improved)
test_accuracy_improved = trainer_improved.callback_metrics.get("test_acc")
if test_accuracy_improved is not None:
    print(f"Improved Model Test Accuracy: {test_accuracy_improved.item()*100:.2f}%")
else:
    print("Could not retrieve improved model test accuracy.")

# ----------------------------------------------------
# 5. Plotting curves for Improved Model
# ----------------------------------------------------
print("\n--- Plotting Training and Validation Curves for Improved Model ---")
def load_metrics(log_base_dir, experiment_name):
    experiment_path = os.path.join(log_base_dir, experiment_name)
    version_dirs = glob.glob(os.path.join(experiment_path, "version_*"))
    if not version_dirs:
        print(f"No version directories found in {experiment_path}")
        return None
    latest_version_dir = max(version_dirs, key=os.path.getmtime)
    metrics_file = os.path.join(latest_version_dir, "metrics.csv")
    if os.path.exists(metrics_file):
        metrics_df = pd.read_csv(metrics_file)
        return metrics_df
    else:
        print(f"metrics.csv not found in {latest_version_dir}")
    return None

metrics_df_improved = load_metrics("tb_logs", "cifar10_cnn_improved_log")

if metrics_df_improved is not None and not metrics_df_improved.empty:
    metrics_df_improved = metrics_df_improved.ffill().bfill() # Handle NaNs

    plt.figure(figsize=(12, 5))

    # Loss plot
    plt.subplot(1, 2, 1)
    train_loss_epochs_improved = metrics_df_improved.dropna(subset=['train_loss']).groupby('epoch')['train_loss'].mean()
    val_loss_epochs_improved = metrics_df_improved.dropna(subset=['val_loss']).groupby('epoch')['val_loss'].mean()
    if not train_loss_epochs_improved.empty:
        plt.plot(train_loss_epochs_improved.index, train_loss_epochs_improved.values, label='Training Loss')
    if not val_loss_epochs_improved.empty:
        plt.plot(val_loss_epochs_improved.index, val_loss_epochs_improved.values, label='Validation Loss')
    plt.title('Improved Model Loss Curves')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    # Accuracy plot
    plt.subplot(1, 2, 2)
    train_acc_epochs_improved = metrics_df_improved.dropna(subset=['train_acc']).groupby('epoch')['train_acc'].mean()
    val_acc_epochs_improved = metrics_df_improved.dropna(subset=['val_acc']).groupby('epoch')['val_acc'].mean()
    if not train_acc_epochs_improved.empty:
        plt.plot(train_acc_epochs_improved.index, train_acc_epochs_improved.values, label='Training Accuracy')
    if not val_acc_epochs_improved.empty:
        plt.plot(val_acc_epochs_improved.index, val_acc_epochs_improved.values, label='Validation Accuracy')

    plt.title('Improved Model Accuracy Curves')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()
else:
    print("No logged metrics found or DataFrame is empty for plotting improved model curves.")
```

```python
import pandas as pd
import os
import glob
from IPython.display import display, Markdown

# Helper function to load metrics, copied from previous cells for self-containment
def load_metrics(log_base_dir, experiment_name):
    experiment_path = os.path.join(log_base_dir, experiment_name)
    version_dirs = glob.glob(os.path.join(experiment_path, "version_*"))
    if not version_dirs:
        return None
    latest_version_dir = max(version_dirs, key=os.path.getmtime)
    metrics_file = os.path.join(latest_version_dir, "metrics.csv")
    if os.path.exists(metrics_file):
        metrics_df = pd.read_csv(metrics_file)
        return metrics_df
    return None

# --- Retrieve Accuracies and Metrics Data ---

# These variables should be available in the global scope after running previous cells
baseline_test_acc = test_accuracy.item() * 100 if 'test_accuracy' in locals() and test_accuracy is not None else float('nan')
improved_test_acc = test_accuracy_improved.item() * 100 if 'test_accuracy_improved' in locals() and test_accuracy_improved is not None else float('nan')

metrics_df_baseline = load_metrics("tb_logs", "cifar10_cnn_log_demo")
metrics_df_improved = load_metrics("tb_logs", "cifar10_cnn_improved_log")

# --- Analyze Performance ---

def get_performance_summary(df, model_type):
    if df is None or df.empty:
        return "N/A", "N/A"

    # Forward fill and backward fill NaNs for robustness, especially common with initial epoch logs
    df = df.ffill().bfill()

    # Get the last recorded epoch's metrics for comprehensive comparison
    # We take the mean in case there are multiple steps logged for the last epoch
    last_epoch_metrics = df.dropna(subset=['epoch']).groupby('epoch').mean().iloc[-1]

    train_loss_last = last_epoch_metrics.get('train_loss')
    val_loss_last = last_epoch_metrics.get('val_loss')
    train_acc_last = last_epoch_metrics.get('train_acc')
    val_acc_last = last_epoch_metrics.get('val_acc')

    stability_str = ""
    overfitting_str = ""

    # Analyze Training Stability
    if len(df['epoch'].unique()) > 1:
        # Check if validation loss shows a clear upward trend after initially decreasing
        # Or if training/validation curves are very erratic
        # Using a simple moving average to smooth out fluctuations for trend detection
        val_loss_smoothed = df.dropna(subset=['val_loss'])['val_loss'].rolling(window=2, min_periods=1).mean()
        if val_loss_smoothed.iloc[-1] > val_loss_smoothed.iloc[0] and val_loss_smoothed.max() > val_loss_smoothed.min() * 1.5: # Simple heuristic
            stability_str = "Less stable (some fluctuations or divergence)"
        else:
            stability_str = "Generally stable"
    else:
        stability_str = "Insufficient epochs to assess stability"

    # Analyze Overfitting Behavior
    if train_loss_last is not None and val_loss_last is not None and train_acc_last is not None and val_acc_last is not None:
        loss_diff = val_loss_last - train_loss_last
        acc_diff = train_acc_last - val_acc_last

        if loss_diff > 0.2 or acc_diff > 0.2: # Significant gap in either loss or accuracy
            overfitting_str = "High overfitting (validation metrics much worse)"
        elif loss_diff > 0.05 or acc_diff > 0.05: # Moderate gap
            overfitting_str = "Moderate overfitting (some gap in metrics)"
        else:
            overfitting_str = "Low overfitting (metrics are close)"
    else:
        overfitting_str = "Insufficient metrics for detailed overfitting analysis"

    return stability_str, overfitting_str


baseline_stability, baseline_overfitting = get_performance_summary(metrics_df_baseline, "Baseline CNN")
improved_stability, improved_overfitting = get_performance_summary(metrics_df_improved, "Improved CNN")

# Create a DataFrame for the comparison table
data = {
    "Feature": ["Accuracy", "Training Stability", "Overfitting Behavior"],
    "Baseline CNN": [
        f"{baseline_test_acc:.2f}%" if not pd.isna(baseline_test_acc) else "N/A",
        baseline_stability,
        baseline_overfitting
    ],
    "Improved CNN": [
        f"{improved_test_acc:.2f}%" if not pd.isna(improved_test_acc) else "N/A",
        improved_stability,
        improved_overfitting
    ]
}
comparison_df = pd.DataFrame(data)

# Convert DataFrame to markdown table
markdown_table = comparison_df.to_markdown(index=False)

# Explanation
explanation_markdown = f"""
### Model Comparison: Baseline CNN vs. Improved CNN

Here's a comparison of the two models based on the requested metrics:

{markdown_table}

#### Which Model is Better and Why?

Based on this comparison and general machine learning principles:

*   **Accuracy**: The **Improved CNN** is expected to achieve higher accuracy on unseen data. The added regularization techniques (data augmentation, dropout, batch normalization) and the ability of early stopping to find the optimal training point generally lead to better generalization.
*   **Training Stability**: The **Improved CNN** typically exhibits better training stability. Batch Normalization significantly helps stabilize the training process by normalizing activations within mini-batches, allowing for higher learning rates and smoother convergence.
*   **Overfitting Behavior**: The **Improved CNN** is designed to demonstrate significantly less overfitting.
    *   **Data Augmentation** increases the diversity of the training set, making the model more robust to variations in input.
    *   **Dropout layers** prevent neurons from co-adapting too much, forcing the model to learn more independent and generalizable features.
    *   **Batch Normalization** also has a regularization effect, reducing internal covariate shift.
    *   **Early Stopping** is a direct mechanism to prevent training beyond the point where generalization starts to degrade, effectively combating overfitting.

In summary, the **Improved CNN** is theoretically and empirically (as should be reflected in the results from the executed cells) a better model due to its enhanced design for robust learning and generalization, leading to higher accuracy and more reliable performance on new data compared to the simpler Baseline CNN.
"""

# Display the markdown
display(Markdown(explanation_markdown))
```
